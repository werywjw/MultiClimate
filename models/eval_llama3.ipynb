```python
import ollama
import concurrent.futures
import pandas as pd
```

```python
model_name = 'llama3'
```

```python
def process_prompt(prompt):
    return ollama.generate(model=model_name, prompt=prompt) 
```

```python
video = 'WICC'
```

```python
data = pd.read_csv(f'dataset/{video}/{video}.csv', header=None, skiprows=1, names=['label', 'text']) 
text = data['text']
```

```python
def predict_stance(text):
    if not text.strip():
        return 0 

    prompt = f"Classify the stance of the following text towards climate change as: 0 (neutral), 1 (support), 2 (oppose): '{text}'"
    response = ollama.generate(model=model_name, prompt=prompt)
    print(response)
    prediction = response['response'].strip()
    
    if "1" in prediction:
        return 1
    elif "2" in prediction:
        return 2
    else:
        return 0 
```

```python
data['predicted_label'] = data['text'].apply(predict_stance)
```
